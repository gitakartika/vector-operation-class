```python
#Created By: Gita Kartika Suriah

# In[ ]:

import numpy
import math
class Vector(object):
    def __init__(self,*v):
        self.values= v
#Vector Addition
    def __add__(self,vector):
        total=tuple(a+b for a,b in zip(self.values,vector.values))
        return vector(*total)
#Vector Substraction
    def __sub__(self, vector):
        total = tuple( a - b for a, b in zip(self.values, vector.values) )
        return vector(*total)
#Scalar Multiplication
    def __rmul__(self,scalar):
        vector= tuple((scalar*numpy.array(self.values)).tolist())
        return vector(*vector)
#Metric
    def metric(self,vector):
        x=tuple((a-b)**2 for a, b in zip(self.values, vector.values))
        y=[]
        for i in range (1,len(vector.values)+1):
            y.append(i)
        return math.sqrt(sum(a*b for a,b in zip(x,y)))
#Norm Vector
    def norm(self):
        return math.sqrt(sum(x**2 for x in self.values))
#Vector Inner Product
    def ip(self, vector):
        x=tuple(a*b for a, b in zip(self.values,vector.values))
        y=[]
        for i in range (1,len(vector.values)+1):
            y.append(i)
        return sum(a*b for a,b in zip(x,y))
#Print Vector
    def __repr__(self):
        return str(self.values)

# In[ ]:

a=vector(4,1,3)
b=vector(1,-1,1)
c=vector(2,4,0)
x=vector(1,0,2,0)
y=vector(3,2.5,0,1.5)
z=vector(0,0,3,1.5)

d=a+b
print("d:",d)
e=2*c-b
print("e:",e)
w=3*y-2*(x+z)
print("w:",w)
print("d(a,b):",a.metric(b))
print("d(d,a):",d.metric(a))
print("d(e,c):",e.metric(c))
print("d(x,y):",x.metric(y))
print("d(w,z):",w.metric(z))
print("Norm c:",c.norm())
print("Norm d:",d.norm())
print("Norm e:",e.norm())
print("Norm x:",x.norm())
print("Norm w:",w.norm())
print("<a,c>:",a.ip(c))
print("<d,b>:",d.ip(b))
print("<y,z>:",y.ip(z))
print("<x,w>:",x.ip(w))
```
